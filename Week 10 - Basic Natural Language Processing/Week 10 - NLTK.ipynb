```python
from nltk.util import bigrams, trigrams, ngrams
```

Trigrams : tokens of three consecutive written words
Bigrams : tokens of two consecutive written words
Ngrams : tokens of any number of consecutive written words

```python
string = "Another important aspect of urban planning is that the range of urban planning projects include the large-scale master planning of empty sites or Greenfield projects as well as small-scale interventions and refurbishments of existing structures, buildings and public spaces."
quotes_tokens = nltk.word_tokenize(string)
quotes_tokens
```

```python
quote_bigrams = list(nltk.bigrams(quotes_tokens))
quote_bigrams
```

```python
quote_trigrams = list(nltk.trigrams(quotes_tokens))
quote_trigrams
```

```python
quote_ngrams = list(nltk.ngrams(quotes_tokens, 5)) #trial: tokens of 5 written words
quote_ngrams
```

```python
#Stemming
from nltk.stem import PorterStemmer
```

```python
#using PorterStemmer
stemmer = PorterStemmer()
words = ['play', 'playing', 'plays', 'played',
         'playfullness', 'playful']
words2 = ['give', 'giving', 'given', 'gave']
stemmed = [stemmer.stem(word) for word in words2]

for e1,e2 in zip(words2,stemmed):
    print(e1+' ----> '+e2)
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()

stemmed = [lst.stem(word) for word in words2]

for e1,e2 in zip(words2,stemmed):
    print(e1+' ----> '+e2)
```

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')
```

```python
stemmed = [sbst.stem(word) for word in words2]

for e1,e2 in zip(words2,stemmed):
    print(e1+' ----> '+e2)
```

Named Entity Recognition

The process of detecting the named entities such as a movie, a monetary value, organization, location, quantities, person in text

Example: Google's CEO Sundar Pichai introduced the new Pixel at Minnesota Roi Centre Event

Breakdown:

- Google's - Organization
- Sundar Pichai - Person
- Minnesota - Location
- Roi Centre Event - Organization

```python
from nltk import ne_chunk
```

```python
NE_sent = "Google's CEO Sundar Pichai introduced the new Pixel at Minnesota Roi Centre Event"
```

```python
NE_tokens = word_tokenize(NE_sent)
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
```

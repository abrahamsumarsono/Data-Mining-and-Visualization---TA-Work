Extracting data from a CSV file using crawling method

```python
import io
import pandas as pd
import requests
 
URL = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=T10YIE&cosd=2017-04-14&coed=2022-04-14"
resp = requests.get(URL)
if resp.status_code == 200:
   csvtext = resp.text
   csvbuffer = io.StringIO(csvtext) #converting the CSV data
   df = pd.read_csv(csvbuffer)
   print(df)
```

Extracting Weather Data

```python
from bs4 import BeautifulSoup
import requests

#sukolilo surabaya
URL = "https://weather.com/weather/today/l/78ffc0e114446335f82da7692c927f72cb0ad90a9b607fb5759ad01ad1314a3b"
resp = requests.get(URL)
soup = BeautifulSoup(resp.text, "lxml") #lxml is used to parse HTML content
elements = soup.select('span[data-testid="TemperatureValue"][class^="CurrentConditions"]')
print(elements[0].text)
```

```python
#BooksByCategoryScraper.py
from autoscraper import AutoScraper
import pandas as pd
 
#BooksUrlScraper section
ScienceCategoryLink = 'https://books.toscrape.com/catalogue/category/books/science_22/index.html'
WantedList=["https://books.toscrape.com/catalogue/the-grand-design_405/index.html"]
BooksUrlScraper = AutoScraper()
BooksUrlScraper.build(ScienceCategoryLink, wanted_list=WantedList)
```

```python
#BookInfoScraper section
BookPageUrl = "https://books.toscrape.com/catalogue/the-grand-design_405/index.html" 
WantedList=["The Grand Design", "£13.76", "3213b1f13f5f0f7c", "In stock (5 available)"]
 
BookInfoScraper = AutoScraper()
BookInfoScraper.build(BookPageUrl, wanted_list=WantedList)
```

```python
#Scraping info of each book and storing into an excel file
BooksUrlList = BooksUrlScraper.get_result_similar(ScienceCategoryLink) 
BooksInfoList = [] #the place to store the info

for Url in BooksUrlList:
  book_info= BookInfoScraper.get_result_exact(Url)
  BooksInfoList.append(book_info)
df = pd.DataFrame(BooksInfoList, columns =["Book Title", "Price", "UPC", "Availability"])
df.to_excel("BooksInScienceCategory2.xlsx")
```
